```python
import json
import matplotlib.pyplot as plt
```

```python
#Analyse du dataset_relevance
with open("data/comments_with_relevance.json", "r", encoding="utf-8") as f:
        dataset_relevance = json.load(f)
```

```python
scores = []
for comment in dataset_relevance:
    scores.append(int(comment['relevance_score']))     
print(len(scores))
plt.figure()
plt.hist(scores, bins=30, density=True) 
plt.title(f"relevant score for {len(scores)} comments")
plt.xlabel("relevant score")
plt.ylabel("density") 
plt.savefig("plots/hist30", dpi=600)
```
